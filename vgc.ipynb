```python
import requests
import pandas as pd
from google.colab import data_table

response = requests.get("https://api.statsugiri.gg/teams/gen9vgc2023regulationc/today")
response = response.json()
print("# team: {}".format(response["num_teams"]))
print(response["snapshot_date"])
teams = []
for team in response["teams"]:
  teams.append({
      "Rating": team["rating"],
      "Pokemon 1": team["pkmn_team"][0],
      "Pokemon 2": team["pkmn_team"][1],
      "Pokemon 3": team["pkmn_team"][2],
      "Pokemon 4": team["pkmn_team"][3],
      "Pokemon 5": team["pkmn_team"][4],
      "Pokemon 6": team["pkmn_team"][5],
      "Replay date": team["replay_upload_date"],
      "Replay": "https://replay.pokemonshowdown.com/" + team["replay_id"]
  })


data_table.DataTable(pd.DataFrame(teams), include_index=False, num_rows_per_page=20)
```
